```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
NUM_TRADING_DAYS=252
NUM_SIMULATIONS = 100000
stocks = ['RELIANCE.NS','TCS.NS','HINDUNILVR.NS','HDFCBANK.NS','ITC.NS','LT.NS','INFY.NS']
start_date="2018-01-01"
end_date="2024-09-01"
```

```python
stock_data={}
for i in stocks:
  ticker=yf.Ticker(i)
  stock_data[i]=ticker.history(start = start_date, end= end_date,)['Close']
  stock_prices = pd.DataFrame(stock_data)
```

```python
stock_prices
```

log returns are editable, in normal returns u can't edit , use dropna to drop the nan values

```python
log_returns = np.log(stock_prices/stock_prices.shift(1))
```

```python
log_returns = log_returns.dropna()
```

```python
print("Stocks Annualised Returns")
print(log_returns.mean()*NUM_TRADING_DAYS)
```

```python
print("Stocks   Volatility(Risk)")
print(np.std(log_returns)*np.sqrt(NUM_TRADING_DAYS))
```

```python
portfolio_return=[]
portfolio_risk=[]
portfolio_weight=[]
```

```python
for _ in range (NUM_SIMULATIONS):
  w=np.random.random(len(stocks))
  w/=np.sum(w)
  portfolio_weight.append(w)
  p_return=np.sum(log_returns.mean()*w)*NUM_TRADING_DAYS
  portfolio_return.append(p_return)
  p_risk = np.sqrt(np.dot(w.T,
                        np.dot(log_returns.cov()*NUM_TRADING_DAYS,w)))
  portfolio_risk.append(p_risk)
```

```python
portfolio_weight = np.array(portfolio_weight)
portfolio_weight = np.array(portfolio_risk)
portfolio_weight = np.array(portfolio_return)
```

```python
portfolios = pd.DataFrame({
    "Returns": portfolio_return,
    "Risk": portfolio_risk,
    "Sharpe_Ratio": [x / y for x, y in zip(portfolio_return, portfolio_risk)] # Divide corresponding elements from portfolio_return and portfolio_risk using list comprehension
})
```

```python
# Convert lists to NumPy arrays for element-wise division
portfolio_return = np.array(portfolio_return)
portfolio_risk = np.array(portfolio_risk)

plt.figure(figsize=(10,6))
plt.scatter(portfolio_risk,portfolio_return,
            c=portfolio_return/portfolio_risk,marker='o')
plt.grid(True)
plt.xlabel('Expected_Risk')
plt.ylabel('Expected_Return')
plt.colorbar(label="Sharpe Ratio")
plt.title("Expected Returns vs Expected_Risk")
plt.tight_layout()
```

```python
sharpe_ratio=portfolio_return/portfolio_risk
```
